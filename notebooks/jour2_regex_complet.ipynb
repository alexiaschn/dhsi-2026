## expressions régulières de base

[Cheat Sheet](https://cheatography.com/davechild/cheat-sheets/regular-expressions/)

. : n'importe quel caractère

\ : précède un caractère que l'on souhaite "échapper". Ex: re.search("\.", texte) va chercher le premier point du texte tandis que re.search(".", texte) sélectionnera le premier caractère du texte.

### Quelques caractères échappés

\d : un chiffre
\D : non chiffre
\w : une lettre 
\W : non lettre

Ex : re.search("d \d", texte) sélectionne la première instance du motif "d 1" ou "d 2" ou "d 3" etc. 


^ : début d'une chaîne.
\b : frontière de mot
\s : espace
$ : fin de chaîne
\n : saut de ligne
\t : tabulation


() : effectuer une capture ex: re.search(r"date: (\d{4})", "date: 2026") retourne "2026"
| : équivalent de OU  exemple re.search(r"(chien|chat)", "j'aime les chiens") retourne "chien". 
[] : range, correspond toujours à un caractère ex: re.search(r"[A-Z]\w*", "mon nom est Alexia") retourne "Alexia". Exemple 2 : re.search(r"[abc].", "magie") retourne "ag" (a ou b ou c suivi de n'importe quel caractère).


### Les quantifieurs

Suivent un caractère pour quantifier le nombre de fois où il apparait. 

* : 0 ou n fois
+ : 1 ou n fois
? : 0 ou 1 fois

### Module re en Python

[Documentation](https://docs.python.org/3/library/re.html)

re.match(pattern, string) -> match : sélectionne à partir du début de la chaîne
re.search(pattern, string) -> match : sélectionne le motif n'importe où dans la chaîne (s'arrête à la première occurrence)
re.findall(pattern, string) -> list : sélectionne toutes les instances du motif dans la chaîne
re.split(pattern, string) -> list : divise la chaine selon le motif regex 
re.sub(pattern, replacement_pattern, string) -> string : remplace un motif par un autre

match.group(0) : retourne toute la séquence du motif identifiée dans le texte
match.group(1) : retourne le premier groupe capture

\1 = match.group(1)
\2 = match.group(2) etc.

## Exercices plus facile

Chercher toutes les lignes contenant les noms d'auteur. 

Chercher toutes les lignes contenant le mot 'chien' ou 'oiseau'. 

Transformer tous les a en b. 

Transformer tous les a majuscule ou minuscule en b. 

Inverser toutes les lettres a suivies d'un i en ia ex: voulait -> vouliat. 

Capturer les textes écrits par Diogène Laërce


```python
# Chercher toutes les lignes contenant les noms d'auteur. 
print('\n'.join([author for author in re.findall(r'author_fr:.*', texte_original)]))
```

```python
# Chercher toutes les épigrammes contenant un chien ou un oiseau. 
print('\n'.join([chien_chat[0] for chien_chat in re.findall(r'(text_fr:.*(chien|oiseau).*)', texte_original)]))
```

```python
# Transformer tous les a en b.
print(re.sub(r'a', 'b', texte_original))
```

```python
# Inverser toutes les lettres a suivies d'un i en ia ex: voulait -> vouliat. 
print(re.sub(r'ai', 'ia', texte_original))
```

```python
# Capturer les textes écrits par Diogène Laërce

print('\n'.join([epi for epi in re.findall(r'author_fr: Diogène Laërce\ntext_fr: (.*)', texte_original)]))
```

# Nettoyer un corpus 

Nettoyer un corpus comprend plusieurs étapes :
- s'assurer que les données correspondent à ce qu'on attend. Exemple : corriger une OCRisation -> remplacer les caractères qui ont mal été retranscrits.
- uniformiser ses choix. Exemple : utiliser les guillements français «» et non anglais ""
- structurer l'ensemble des documents. Exemple : formatter ses documents en texte brut en CSV (comma separated value)
- normaliser l'ensemble du corpus. Exemple : convention de nommage avec un identifiant unique par document, pas d'espace et de caractères spéciaux pour éviter les problèmes de manipulation. 


Pour l'exemple, imaginons que nous venons de trouver dans des archives 3 nouvelles épigrammes et que nous souhaitons les ajouter à notre corpus en CSV. 

Nous avons OCRisé les 3 épigrammes et le résultat est le suivant :
urn:cts:greekLit:tlg7000.tlg001.ag:5.17   Gardienne du rivage battu des flotsƒ, je t’envoie Ces gâteaux et ces présents d’un simple sacrifice.
```python
# Télécharger le corpus 'epigrammes_a_nettoyer.txt' et le lire


try:
    with open(f"{os.getcwd()}/corpus/epigrammes_a_nettoyer.txt", 'r', encoding='utf-8') as f:
        a_corriger = f.read()
except FileNotFoundError:
    !wget -P corpus https://raw.githubusercontent.com/alexiaschn/dhsi-2026/main/notebooks/corpus/epigrammes_anthologia_graeca_fr.txt
    with open(f"{os.getcwd()}/corpus/epigrammes_a_nettoyer.txt", 'r', encoding='utf-8') as f:
        a_corriger = f.read()

print(a_corriger)
```

```python
# identifier les problèmes de normalisation à corriger sur les épigrammes
# supprimer les caractères qui ne devraient pas être là ƒ, €, 
texte_propre = re.sub(r'[ƒ€]', '', a_corriger )
print(texte_propre)
```

```python
# Nous allons structurer en liste chaque ligne avec un split()
texte_propre = texte_propre.split('\n')
print('\n'.join(texte_propre))
```

```python
# Structurons les données en récupérant l'identifiant de l'épigramme, le livre et le numéro de fragment qui se trouve à la fin de l'URI (ex: urn:cts:greekLit:tlg7000.tlg001.ag:5.17 rend livre 5 et fragment 17)
# nous avons besoin des valeurs du tableau : epigramme_id,book,fragment,urn,author_fr,text_fr
corpus_a_ajouter = []
for texte in texte_propre:
    urn = re.search(r'urn:cts:greekLit:tlg7000\.tlg001\.ag:\d\.\d+', texte)
    urn = urn.group()
    identifiant = re.search(r'(\d\.\d+)\W', texte)
    identifiant = identifiant.group(1)
    livre, fragment = identifiant.split('.')
    text_fr = re.sub(urn, '', texte)
    author = re.search(r'---(.*)', texte)
    text_fr = re.sub(author.group(0), '', text_fr)
    author = author.group(1)
    corpus_a_ajouter.append({'epigramme_id': identifiant,
                            'book': livre, 
                            'fragment': fragment, 
                            'urn': urn, 
                            'author_fr': author,
                            'text_fr': text_fr
                            })

print(corpus_a_ajouter)
```

```python
# On va nettoyer le texte text_fr 
for epigramme in corpus_a_ajouter:
    # suppression des espaces en tête de texte et des guillemets inutiles
    epigramme['text_fr'] = re.sub(r'^\W+', '', epigramme['text_fr'])
    epigramme['text_fr'] = re.sub(r'\W+$', '', epigramme['text_fr'])
    print(epigramme['text_fr'])
    
```

```python
# vérifions que tout est propre
for epigramme in corpus_a_ajouter: 
    print(epigramme)
```

```python
# intégrer les nouveaux épigrammes à notre corpus existants en csv 
import csv

with open(f"{os.getcwd()}/corpus/epigrammes_anthologia_graeca_fr.csv", 'a', encoding='utf-8') as g:
    writer = csv.DictWriter(g, fieldnames=['epigramme_id','book','fragment','urn','author_fr','text_fr'])
    writer.writerows(corpus_a_ajouter)
```
