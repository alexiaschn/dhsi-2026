## expressions régulières de base

[Cheat Sheet](https://cheatography.com/davechild/cheat-sheets/regular-expressions/)

. : n'importe quel caractère

\ : précède un caractère que l'on souhaite "échapper". Ex: re.search("\.", texte) va chercher le premier point du texte tandis que re.search(".", texte) sélectionnera le premier caractère du texte.

### Quelques caractères échappés

\d : un chiffre
\D : non chiffre
\w : une lettre 
\W : non lettre

Ex : re.search("d \d", texte) sélectionne la première instance du motif "d 1" ou "d 2" ou "d 3" etc. 


^ : début d'une chaîne.
\b : frontière de mot
\s : espace
$ : fin de chaîne
\n : saut de ligne
\t : tabulation


() : effectuer une capture ex: re.search(r"date: (\d{4})", "date: 2026") retourne "2026"
| : équivalent de OU  exemple re.search(r"(chien|chat)", "j'aime les chiens") retourne "chien". 
[] : range, correspond toujours à un caractère ex: re.search(r"[A-Z]\w*", "mon nom est Alexia") retourne "Alexia". Exemple 2 : re.search(r"[abc].", "magie") retourne "ag" (a ou b ou c suivi de n'importe quel caractère).


### Les quantifieurs

Suivent un caractère pour quantifier le nombre de fois où il apparait. 

* : 0 ou n fois
+ : 1 ou n fois
? : 0 ou 1 fois

### Module re en Python

[Documentation](https://docs.python.org/3/library/re.html)

re.match(pattern, string) -> match : sélectionne à partir du début de la chaîne
re.search(pattern, string) -> match : sélectionne le motif n'importe où dans la chaîne (s'arrête à la première occurrence)
re.findall(pattern, string) -> list : sélectionne toutes les instances du motif dans la chaîne
re.split(pattern, string) -> list : divise la chaine selon le motif regex 
re.sub(pattern, replacement_pattern, string) -> string : remplace un motif par un autre

match.group(0) : retourne toute la séquence du motif identifiée dans le texte
match.group(1) : retourne le premier groupe capture

\1 = match.group(1)
\2 = match.group(2) etc.

```python
# Chercher toutes les épigrammes contenant un chien ou un chat. 
print(re.findall(r'(text_fr:.*(chien|chat).*)', texte_original))
```

```python
# Transformer tous les a en b.
print(re.sub(r'a', 'b', texte_original))
```

```python
# Inverser toutes les lettres a suivies d'un i en ia ex: voulait -> vouliat. 
print(re.sub(r'ai', 'ia', texte_original))
```

```python
# Capturer les textes écrits par Xénophane

print(re.findall(r'author_fr: Xénophane\ntext_fr: (.*)', texte_original))
```

```python

# Extraire tous les numéros de fragment. 
numeros = re.findall(r"fragment: (\d{3})", texte_original)
print(f"Numéros des fragments : {numeros}")
```

```python
# Extraire toutes les épigrammes avec leurs métadonnées. 
epigrammes_completes = re.split(r"\n\n", texte_original)
print(f"Épigrammes avec leurs métadonnées : {epigrammes_completes}")
```

```python
# Extraire le premier texte. 

premier_texte = re.search(r"text_fr:(.*)", texte_original)
print(f"Premier texte : {premier_texte.group(1)}")
```

```python
# Extraire le dernier texte

dernier_texte = re.search(r".*text_fr:(.*)", texte_original, flags=re.DOTALL)
print(f"Premier texte : {dernier_texte.group(1)}")
```

```python
# Numéro de l'épigramme qui contient le mot "rossignol"

oiseau = re.search(r"fragment: (\d{3})(?:(?!epigramme_id:).)*rossignol", texte_original, flags=re.DOTALL)
print(f"Oiseau fragment id : {oiseau.group(1)}")
```

```python
# Substitution de tous les noms d'auteur commençant par D par votre nom. 

x_nom = re.sub(r"(author_fr: )([D].*)", r"\1Alexia", texte_original )
print(x_nom)
```

```python
# Nettoyer le texte

def nettoyer_texte(texte: str) -> str:
    """Nettoie le texte français récupéré depuis l'API."""
    # Recolle les mots coupés par un tiret suivi d'un espace ou d'un retour à la ligne.
    # Exemple : "sau- terelle" devient "sauterelle".
    texte = re.sub(r"(?<=\w)-\s+(?=\w)", "", texte)

    # Transforme ensuite les retours à la ligne et espaces multiples en espaces simples.
    return re.sub(r"\s+", " ", texte).strip()

for epigramme in corpus:
    epigramme['texte'] = nettoyer_texte(epigramme['texte'])
```
